```python
#Pattern series in python
#1.
n= int(input("Enter the no of lines:"))
for i in range(n//2):
    print("*",end="")
    for space in range(0,i):
        print("*",end="")
    print("\n")
for i in range(n//2,-1,-1):
    print("*",end="")
    for space in range(i):
        print("*",end="")
    print("\n")
```

```python
#2.
n=int(input("Enter no of lines:"))
for i in range(n//2,-1,-1):
    for j in range(0,i):
        print(" ",end="")
    for star in range(n//2,i,-1):
        print("*",end="")
    print("\n")
for i in range(0,n//2):
    for j in range(-1,i):
        print(" ",end="")
    for star in range(i,n//2-1):
        print("*",end="")
    print("\n")
```

```python
#2.
n=int(input("Enter the no of lines:"))
for i in range(n,-1,-1):
    for j in range(0,i):
        print(" ",end="")
    for star in range(n,i,-1):
        print("*",end="")
    print("\n")
```

```python
#4.
n=int(input("Enter the no of lines:"))
for i in range(n//2,-1,-1):
    for j in range(0,i):
        print(" ",end="")
    for star in range(0,n//2):
        for k in range(0,star):
            print("*",end="")
    print("\n")
for i in range(0,n//2+1):
    for j in range(0,i):
        print(" ",end="")
    for star in range(0,n//2):
        for k in range(0,star):
            print("*",end="")
    print("\n")
```

```python
#5.
n=int(input("Enter the no of lines :"))
for i in range(n//2,0,-1):
    for j in range(0,i):
        print(" ",end="")
    for k in range(0,i):
        print("*",end="")
    print("\n")
for i in range(0,n//2-1):
    for j in range(0,i+2):
        print(" ",end="")
    for k in range(0,i+2):
        print("*",end="")
    print("\n")
```

```python
#6.
n=int(input("Enter the no of lines:"))
for i in range(n,-1,-1):
    for j in range(0,i):
        print(" ",end="")
    for star in range(2*n-1,2*i,-1):
        print("*",end="")
    print("\n")
```
